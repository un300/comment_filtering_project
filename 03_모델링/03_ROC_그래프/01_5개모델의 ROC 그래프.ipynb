```
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
cnn = pd.read_csv(path[0])
lstm = pd.read_csv(path[1])
cnn_lstm = pd.read_csv(path[2])
lightGBM = pd.read_csv(path[3])
naive_bayes = pd.read_csv(path[4])
```
