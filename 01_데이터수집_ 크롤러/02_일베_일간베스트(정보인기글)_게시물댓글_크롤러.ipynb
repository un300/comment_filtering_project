## 2. "일간베스트" 사이트의 일간베스트 게시물 크롤링 진행

```python
# 페이지 지정 : 세션이 끊기는걸 방지하기 위해 한페이지씩 입력하여 크롤링 해주세요
input_page = [12]

# 전체댓글을 담을 리스트
post_comment = []

for page_num in tqdm_notebook(input_page) :
    url = "https://www.ilbe.com/list/ilbe?page=" + str(page_num) + "&sub=best&listStyle=list"
    
    driver.get(url)
        
    ilbe_bestpost_body_list = driver.find_elements_by_css_selector('#content-wrap > div.board-wrap > div.board-list > ul.board-body > li')

    n = len(ilbe_bestpost_body_list)


    k=1

    for idx in tqdm_notebook(range(5, n)) :

        print(k, '번째 게시물입니다')
        k += 1

        # 게시물을 차례대로 눌러 페이지를 변환
        post = ilbe_bestpost_body_list[idx]
        post_click = post.find_element_by_css_selector('span.title > a')
        post_click.send_keys('\n')

        time.sleep(5)

        # 게시물의 댓글 리스트를 넘겨준다
        page = driver.find_elements_by_css_selector('div.post-comment-wrap > div.comment-group > div.paginate > a')
        page_change = page[2 : (len(page)-2) ]

        m = len(page_change)


        for idx2 in range(m):
            print(idx2 + 1, '페이지')
            num_btn = page_change[idx2]
            try:
                num_btn.send_keys('\n')
            except:
                print('댓글 페이지가 하나입니다.')
            time.sleep(1)


            comment_list = driver.find_elements_by_css_selector('#content-wrap > div.board-wrap > div.board-view > div.post-comment-wrap > div.comment-group > div.comment-item')
            l = len(comment_list)

            # 댓글 크롤링!
            for idx3 in range(l):
                comment = comment_list[idx3]

                try :
                    comment_text = comment.find_element_by_css_selector('div.comment-item-box > div.comment > span.comment-box').text
                    post_comment.append(comment_text)

                except :
                    print('삭제된 댓글 입니다.')

                comment_list = driver.find_elements_by_css_selector('#content-wrap > div.board-wrap > div.board-view > div.post-comment-wrap > div.comment-group > div.comment-item')


            page = driver.find_elements_by_css_selector('div.post-comment-wrap > div.comment-group > div.paginate > a')
            page_change = page[2 : (len(page)-2) ]


        # 게시물을 담은 바디리스트 초기화되기 때문에 재지정
        driver.back()

        time.sleep(3)
        ilbe_bestpost_body_list = driver.find_elements_by_css_selector('#content-wrap > div.board-wrap > div.board-list > ul.board-body > li')
        time.sleep(4)

        
        print('현재까지의 댓글수 :', len(post_comment))
        print('')
        
        
```
