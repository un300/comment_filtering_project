```python
from selenium import webdriver
from bs4 import BeautifulSoup
from tqdm import tqdm_notebook

import time
import pandas as pd
import numpy as np
```

## 1. Selenium으로 일베접속

```python
path='./chrome_Driver/chromedriver.exe'
driver = webdriver.Chrome(path)
driver.get("https://www.ilbe.com/list/celeb")

time.sleep(5)
```

## 2. 일간베스트 - '걸그룹/연예인' 게시판 제목 크롤링하기

```python


ilbe_title_list = []

for page in tqdm_notebook(range(1, 100)):
    print('현재 페이지 :', page)
    
    url = 'https://www.ilbe.com/list/celeb?page=' + str(page) + '&listStyle=list'
    driver.get(url)
    
    time.sleep(1)
    
    title_body_list = driver.find_elements_by_css_selector('body > div > #container > #content-wrap > div.board-wrap > div.board-list > ul.board-body > li')
    
    for title_list in title_body_list[7:]:
        
        try:
            title = title_list.find_element_by_css_selector('span.title > a.subject').text
            ilbe_title_list.append(title)
            
        except:
            continue
    
    time.sleep(4)
            
    print('현재까지 크롤링한 데이터수 :', len(ilbe_title_list))
```

## 3. 파일저장

```python
df = pd.DataFrame({'댓글' : ilbe_title_list})

df.to_csv('./크롤링한 댓글 데이터/일베_걸그룹연예인게시판_타이틀.csv', encoding='utf-8-sig')
```
